```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('results.csv')
df.head()
```

```python
agg = df.groupby(['Protocol', 'Algorithm', 'Length']).mean()
agg = agg.reset_index(level='Length')
agg.head()
```

```python
protocols = agg.index.get_level_values('Protocol').unique()

n_protocols = len(protocols)

fig, axes = plt.subplots(
    nrows=n_protocols,
    ncols=2,
    figsize=(12, 6 * n_protocols),
    squeeze=False
)

for i, protocol in enumerate(protocols):

    protocol_df = agg.loc[protocol]

    x = 'Length'
    hue = protocol_df.index.get_level_values('Algorithm')

    # Comparisons
    sns.lineplot(
        ax=axes[i, 0],
        data=protocol_df,
        x=x,
        y='Comparisons',
        hue=hue
    )

    axes[i, 0].set_title(f'{protocol}: Comparisons')
    axes[i, 0].set_xlabel('Length')
    axes[i, 0].set_ylabel('Comparisons')

    # Time
    sns.lineplot(
        ax=axes[i, 1],
        data=protocol_df,
        x=x,
        y='Time_ns',
        hue=hue
    )

    axes[i, 1].set_title(f'{protocol}: Time')
    axes[i, 1].set_xlabel('Length')
    axes[i, 1].set_ylabel('Time (ns)')

# plt.tight_layout()
plt.savefig('plot.png')
plt.show()
```
